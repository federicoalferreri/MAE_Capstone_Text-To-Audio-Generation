```python
import torch
import torchaudio
import torchaudio.functional as F
import torchaudio.transforms as T
import os

print(torch.__version__)
print(torchaudio.__version__)
```

```python

# Parameters

original_dir='path_folder_original_audio_files'
resampled_dir='path_folder_resampled_audio_files'

# Create directory if it doesn't exist

os.makedirs(resampled_dir, exist_ok=True)

sample_rate = 16000
resample_rate = 48000  # New sample rate

# Iterate over each file in the input directory
for filename in os.listdir(original_dir):
  if filename.endswith('.wav'):  # Check if the file is an audio file
        file_path = os.path.join(original_dir, filename)

        # Load the audio file
        audio_file, sample_rate = torchaudio.load(file_path)

        #Resample audio
        resampled_audio = F.resample(audio_file, orig_freq=sample_rate, new_freq=resample_rate, lowpass_filter_width=64)

        # Save the resampled audio
        resampled_file_path = os.path.join(resampled_dir, f'resampled_{filename}')
        torchaudio.save(resampled_file_path, resampled_audio, resample_rate)


        #print(f'Resampled audio saved to {resampled_file_path}')


print (f'All resampled audio files saved correctly')





```
